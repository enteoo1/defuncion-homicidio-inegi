```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 300)

# INGEGI Original file cleanup
# read csv file

file_name = '../dirty-data/INEGI_exporta_1_7_2024_22_19_13_base_ok.csv'
# the number of row in each data frame
chunksize = 1000

# the list that contains all the dataframes
list_of_dataframes = []

for df in pd.read_csv(file_name, encoding='latin1',dtype='object', chunksize=chunksize):
    # process your data frame here
    # then add the current data frame into the list
    list_of_dataframes.append(df)

# all the dataframes together
dirty = pd.concat(list_of_dataframes)

for i in range(len(list_of_dataframes)):
    list_of_dataframes[i] = None

list_of_dataframes = None
# display DataFrame
dirty
```

```python
#split first column to get 'entity/state' and 'municipality' in different columns
dirty[['IdEstado', 'IdMunicipio']] = dirty['Id'].str.split(' ', n=1, expand=True)
#drop unused columns
dirty = dirty.drop(columns=['Id']);
dirty.head(300)
```

```python
#Replace NaN values with '0'
dirty.loc[dirty['IdEstado'].isnull(), 'IdEstado'] = '0'
dirty.loc[dirty['IdMunicipio'].isnull(), 'IdMunicipio'] = '0'
dirty.loc[dirty['Total'].isnull(), 'Total'] = '0'
dirty.loc[dirty['Hombre'].isnull(), 'Hombre'] = '0'
dirty.loc[dirty['Mujer'].isnull(), 'Mujer'] = '0'
dirty.loc[dirty['No especificado'].isnull(), 'No especificado'] = '0'
dirty['Total'] = dirty['Total'].str.replace(',', '')
dirty['Hombre'] = dirty['Hombre'].str.replace(',', '')
dirty['Mujer'] = dirty['Mujer'].str.replace(',', '')
dirty['No especificado'] = dirty['No especificado'].str.replace(',', '')
dirty['Año'] = dirty['Año'].str.replace(',', '')
dirty = dirty.astype({'Total': 'int64', 
                      'Hombre': 'int64', 
                      'Mujer': 'int64', 
                      'No especificado': 'int64', 
                      'Año': 'int64',
                      'Mes': 'string',
                      'Nombre': 'string',
                      'IdEstado': 'string',
                      'IdMunicipio': 'string'
        })

#dirty['No especificado'] =  pd.to_numeric(dirty['No especificado'], downcast='integer')
dirty.dtypes
#dirty.head(300)
```

```python
#Pad ids with '0'
dirty['IdEstado'] =  dirty['IdEstado'].str.zfill(2)
dirty['IdMunicipio'] = dirty['IdMunicipio'].str.zfill(3)
dirty.head(300)
```

```python
#Add new Column for estado
dirty['Estado'] = np.nan
dirty.rename(columns={'Nombre': 'Municipio'}, inplace=True)
dirty
```

```python
#rearrange columns an add column for State
dirty = dirty.reindex(columns=['IdEstado','Estado', 'IdMunicipio','Municipio','Año','Mes','Hombre','Mujer','No especificado','Total'])
dirty
```

```python
#dirty.loc[dirty["Year"] == 'Total', "State"] = dirty["Municipality"]
df2 = pd.DataFrame()
df2['IdEstado'] = dirty.loc[ (dirty['Mes'] == 'Total') & (dirty['IdEstado'] != '00') & (dirty['IdMunicipio'] == '000'), 'IdEstado']
df2['Estado'] = dirty.loc[ (dirty['Mes'] == 'Total') & (dirty['IdEstado'] != '00') & (dirty['IdMunicipio'] == '000'), 'Municipio']
#& (str(dirty['Month'])=='Total') & (str(dirty['StateId'])!='00')
#state_dic = dirty.loc[dirty['Year']=='Total', 'StateId']
df2.drop_duplicates(inplace=True)

df2.head(300)




#dirty.drop(dirty.loc[dirty['Year']==0].index, inplace=True)
#dirty.head(300)
```

```python
updated = dirty.merge(df2, how='left', on=['IdEstado'], suffixes=('', '_new'))
updated['Estado'] = np.where(pd.notnull(updated['Estado_new']), updated['Estado_new'], updated['Estado'])
updated.drop('Estado_new', axis=1, inplace=True)

dirty = updated 
dirty
```

```python
#drop rows with 'Totals' as we won't use them
dirty.drop(dirty.loc[dirty['Año']=='Total'].index, inplace=True)
dirty.drop(dirty.loc[dirty['Mes']=='Total'].index, inplace=True)
dirty.drop(dirty.loc[dirty['Municipio']=='Total'].index, inplace=True)
dirty.drop(dirty.loc[dirty['IdMunicipio']=='000'].index, inplace=True)
dirty.drop(dirty.loc[dirty['IdEstado']=='00'].index, inplace=True)
dirty
```

```python
dirty.to_csv('../data/clean.csv', index=False) 
```
